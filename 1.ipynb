```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
# Load your time series data into a DataFrame
data = pd.read_csv('Informer2020/data/ETT/WTH.csv')
```

```python
date_column_name = 'date'  # Replace with the actual column name
data_column_name = 'OT'  # Replace with the actual column name

date = pd.to_datetime(data[date_column_name])
data_column = data[data_column_name]
```

```python
# Downsample data by taking every nth data point (adjust 'n' as needed)
n = 10
date = date[::n]
data_column = data_column[::n]
```

```python
# Normalize the data using Min-Max scaling
scaler = MinMaxScaler()
data_column_normalized = scaler.fit_transform(data_column.values.reshape(-1, 1))
```

```python
# Create the plot
plt.plot(date, data_column_normalized.flatten()) 
# Add labels and title
plt.xlabel(date_column_name)
plt.ylabel(data_column_name)
plt.title("OT vs. Date")
plt.savefig('Weather OT.png')
# Show the plot
plt.show()
```
